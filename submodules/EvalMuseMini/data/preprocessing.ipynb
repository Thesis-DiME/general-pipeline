```python
included_dirs = [
    'PixArt-alpha',
    'SD_v2.1',
    'SD_v1.5',
    'SD3',
    'PixArt-Sigma',
    'SDXL-Turbo',
    'SDXL',
]
```

```python
import shutil

dir_path = './dataset/images'

for target_dir in os.listdir(dir_path):
    if target_dir not in included_dirs:
        # Remove the directory
        target_dir = os.path.join(dir_path, target_dir)
        if os.path.exists(target_dir):
            shutil.rmtree(target_dir)
            print(f"Directory '{target_dir}' has been removed.")
        else:
            print(f"Directory '{target_dir}' does not exist.")
    else:
        print(f"Directory '{target_dir}' is in the included list and will not be removed.")
```

```python
import json

with open('./dataset/test.json', 'r') as f:
    data = json.load(f)


new_json = []

def check_path(path, included_dirs):
    for dir in included_dirs:
        if path.startswith(dir):
            return True
    return False

for d in data:
    if check_path(d['img_path'], included_dirs):
        new_json.append(d)
        
with open('./dataset/mini_test.json', 'w') as f:
    json.dump(new_json, f, indent=4)
        
```

```python
import json
import os

json_file_path = './dataset/mini_test.json'
images_path = './dataset/images'

with open(json_file_path, 'r') as file:
    data = json.load(file)

image_paths_in_data = set(item['img_path'] for item in data)

for root, dirs, files in os.walk(images_path):
    for file_name in files:
        relative_path = os.path.relpath(os.path.join(root, file_name), images_path)
        
        if relative_path not in image_paths_in_data:
            img_path = os.path.join(images_path, relative_path)
            if os.path.exists(img_path):
                os.remove(img_path)

print("Process completed. All images not referenced in the JSON data have been removed.")
```
